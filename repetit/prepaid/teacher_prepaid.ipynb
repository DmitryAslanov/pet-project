```python
#Импорт библиотек

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import os
from pathlib import Path
pd.set_option('display.max_columns', 90)
```

```python
teacher_info_df = pd.read_feather('../datasets/teachers_info.feather')
```

```python
teacher_info_df.info()
```

**Дроп дублирующихся колонок**

```python
duplicates = {}

# Перебор всех колонок для сравнения
for i, col1 in enumerate(teacher_info_df.columns):
    for col2 in teacher_info_df.columns[i + 1:]:
        # Проверка на идентичность данных в колонках
        if teacher_info_df[col1].equals(teacher_info_df[col2]):
            # Запись найденных дубликатов
            duplicates.setdefault(col1, []).append(col2)

# Вывод результатов
if duplicates:
    for key, values in duplicates.items():
        print(f"Колонка '{key}' дублируется в колонках: {values}")
else:
    print("Дублирующихся колонок не найдено.")
```

```python
teacher_info_df = teacher_info_df.drop(columns=duplicates)
teacher_info_df.info()
```

```python
# Вычисляем процент пропущенных данных для каждой колонки
percent_missing = teacher_info_df.isnull().mean() * 100

# Определяем колонки для удаления
cols_to_drop = percent_missing[percent_missing > 85].index

# Удаляем эти колонки
teacher_info_df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
teacher_info_df = teacher_info_df.drop(columns=['date_update','external_comments.2','information.2','rules_confirmed_date','status_relevant_date.2','partner_id'])
teacher_info_df
```

```python
teacher_info_df['teaching_start_date'] = teacher_info_df['teaching_start_date'].astype(str).str[:10]
```

```python
#Формат даты время
cols_to_convert = ['reg_date', 'birth_date', 'teaching_start_date', 'last_visited', 'relevance_date']

for col in cols_to_convert:
    teacher_info_df[col] = pd.to_datetime(teacher_info_df[col])
```

```python
teacher_info_df
```

```python
current_date = datetime.now()
current_year = current_date.year

# Лет учителю (возраст)
teacher_info_df['teacher_age'] = teacher_info_df['birth_date'].apply(lambda x: (current_date - x).days / 365.25).astype(int)

# Лет преподает
teacher_info_df['years_teaching'] = teacher_info_df['teaching_start_date'].apply(lambda x: (current_date - x).days / 365.25).astype(int)

# Расчет возраста на момент регистрации
teacher_info_df['age_at_reg'] = teacher_info_df['reg_date'].dt.year - teacher_info_df['birth_date'].dt.year

# Расчет опыта преподавания
teacher_info_df['years_teaching'] = (current_date - teacher_info_df['teaching_start_date']).dt.days / 365.25

# Расчет времени с регистрации
teacher_info_df['years_since_reg'] = (current_date - teacher_info_df['reg_date']).dt.days / 365.25

# Расчёи дней без посещения
teacher_info_df['days_since_last_visit'] = (current_date - teacher_info_df['last_visited']).dt.days

# Извлечение года, месяца и дня
teacher_info_df['reg_year'] = teacher_info_df['reg_date'].dt.year
teacher_info_df['reg_month'] = teacher_info_df['reg_date'].dt.month
teacher_info_df['reg_day'] = teacher_info_df['reg_date'].dt.day
teacher_info_df['teaching_start_year'] = teacher_info_df['teaching_start_date'].dt.year
```

```python
# Новый признак есть/нет фото
teacher_info_df['has_photo'] = teacher_info_df['photo_path'].apply(
    lambda x: 0 if x == 'no_photo' or pd.isnull(x) else 1
).fillna(0).astype(int)
```

**Удалим не нужные значаения**

relevance_date удаляем,так как наша модели без разницы время обновления заявки,если ее будут часто запускать(как сказал заказчик)

```python
teacher_info_df = teacher_info_df.drop(columns=['reg_date','birth_date','teaching_start_date','last_visited','photo_path','relevance_date'])
```

```python
teacher_info_df = teacher_info_df.fillna(0)
```

```python
teacher_info_df.describe()
```

```python
# Условия для фильтрации
conditions = (
    (teacher_info_df['lesson_duration.2'] <= 100) &
    (teacher_info_df['teacher_age'] > 16) & (teacher_info_df['teacher_age'] <= 70) &
    (teacher_info_df['years_teaching'] <= 60) &
    (teacher_info_df['age_at_reg'] <= 70)
)

# Применяем условия и фильтруем датафрейм
filtered_df = teacher_info_df[conditions]

# Выводим информацию о количестве строк до и после фильтрации
print(f'Количество строк до фильтрации: {teacher_info_df.shape[0]}')
print(f'Количество строк после фильтрации: {filtered_df.shape[0]}')
```
